```python
import json

train_data = []
with open('train_data.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        train_data.append(json.loads(line))

test_data = []
with open('train_data.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        test_data.append(json.loads(line))

train_df = pd.json_normalize(
    train_data,
    record_path='messages',
    meta=['conversationId', 'respondentWorkerId', 'initiatorWorkerId']
)

test_df = pd.json_normalize(
    test_data,
    record_path='messages',
    meta=['conversationId', 'respondentWorkerId', 'initiatorWorkerId']
)
```

```python
print("=== Basic Numeric Statistics ===")
print(train_df.describe())

print("\n=== Message Length Statistics ===")
print(train_df["messageLength"].describe())

print("\n=== Unique Counts ===")
print(f"Unique Conversations: {train_df['conversationId'].nunique()}")
print(f"Unique Senders: {train_df['senderWorkerId'].nunique()}")
print(f"Unique Respondents: {train_df['respondentWorkerId'].nunique()}")
print(f"Unique Initiators: {train_df['initiatorWorkerId'].nunique()}")

print("\n=== Messages Per Conversation Statistics ===")
print(train_df.groupby("conversationId")["messageId"].count().describe())

print("\n=== Conversation Duration Statistics (based on timeOffset) ===")
duration_per_convo = train_df.groupby("conversationId")["timeOffset"].agg(["min", "max"])
duration_per_convo["duration"] = duration_per_convo["max"] - duration_per_convo["min"]
print(duration_per_convo["duration"].describe())
```
